```python
tide_heights.append(2.0)
```

```python
mean_tide = np.mean(tide_heights)
std_tide = np.std(tide_heights)
tsunami_deviation = (2.0 - mean_tide) / std_tide
```

```python
plt.figure(figsize=(8, 6))
plt.hist(tide_heights, bins=15, color='blue', edgecolor='black', alpha=0.7)
plt.axvline(2.0, color='red', linestyle='--', label='2ft Tsunami Outlier')
plt.axvline(mean_tide, color='green', linestyle='-', label='Mean Tide')
plt.xlabel('Tide Height (ft)')
plt.ylabel('Frequency')
plt.title('Histogram of Tide Heights (Including 2ft Tsunami Outlier)')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
print(f"Mean Tide Height: {mean_tide:.2f} ft")
print(f"Standard Deviation of Tide Heights: {std_tide:.2f} ft")
print(f"The 2ft tsunami is {tsunami_deviation:.2f} standard deviations from the mean.")
```
