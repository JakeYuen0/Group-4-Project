# Group Project Step 1

Importing matplotlib.pyplot and numpy

```python
import matplotlib.pyplot as plt
import numpy as np
```

Importing the data

```python
data = np.genfromtxt("ASTR19_F24_group_project_data.txt", dtype = [('day', 'i8'), ('time', 'U6'), ('height', 'f8')])
```

Assigning values to each column

```python
day = []
time = []
height = []

for i in range(82):
    #first column is days
    day.append(data[i][0])
    #second column is hours: minutes
    time.append(data[i][1])
    #third column is height
    height.append(data[i][2])
```

Formula that combines day with time for the x-axis

```python
x_values = []

for i in range(len(time)):
    # Split hours and minutes from the time string
    hours, minutes = map(int, time[i].split(":"))
    # Calculate fractional day
    fractional_day = day[i] + (hours / 24) + (minutes / 1440)
    x_values.append(fractional_day)
```

Converting the lists into numpy arrays for convience sake

```python
x_values = np.array(x_values)

height = np.array(height)
```

Plotting everything

```python
#adjusting dimensions of the plot
plt.figure(figsize=(10, 6))

#plotting everything
plt.plot(x_values, height, marker='o', linestyle='-', color='blue', label="Tide Height")

#adjusting labels, titles, and adding a grid
plt.xlabel("Day of the Year", fontsize=12)
plt.ylabel("Tide Height (ft)", fontsize=12)
plt.title("Santa Cruz Tide Heights (Jan-Feb 2022)", fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)

#saving the plot
plt.savefig("ASTR_19_Group_Project_Part#1.png")
```
